```
def CorrFunc(f1,i1):
  mean_f=f1["%Change"].mean()
  mean_i=i1["%Change"].mean()
  #print(mean_i)
  a_f=f1["%Change"]-mean_f
  a_i=i1["%Change"]-mean_i
  a_f.reset_index(drop=True, inplace=True)
  a_i.reset_index(drop=True, inplace=True)
  #print(a_f)
  sq_f=a_f**2

  sq_i=a_i**2
  x=a_i.multiply(a_f,axis='index')
  xs=x.sum()
  y=(sq_i.sum()*sq_f.sum())**0.5
  corr=xs/y
  #print("x",x)
  #print("y",y)
  print("corr=",corr)

  return corr
  #print(sq_f)
```

```
first_corr=CorrFunc(f1,i1)
second_corr=CorrFunc(f2,i2)
third_corr=CorrFunc(f3,i3)
fourth_corr=CorrFunc(f4,i4)
fifth_corr=CorrFunc(f5,i5)
```

```
def find_beta(f1,i1):
  std_1=f1["%Change"].std()/i1["%Change"].std()
  beta_1= first_corr*std_1
  return beta_1
```

```
beta_1=find_beta(f1,i1)

beta_2=find_beta(f2,i2)
beta_3=find_beta(f3,i3)
beta_4=find_beta(f4,i4)
beta_5=find_beta(f5,i5)
```

```
def find_sharp(beta,f1):
  sd=f1["%Change"].std()

  sharp=(beta-0.0665)/sd
  print("Sharp's Ratio=",sharp)

  return sharp
```

```
s1=find_sharp(beta_1,f1)
s2=find_sharp(beta_2,f2)
s3=find_sharp(beta_3,f3)
s4=find_sharp(beta_4,f4)
s5=find_sharp(beta_5,f5)
```

```
def find_treynor(f1,beta):
  f_return=(f1["%Change"][-1:])/100

  treynor=(f_return-0.0665)/beta
  print("Treynor's Ratio=",treynor)

  return treynor
```
